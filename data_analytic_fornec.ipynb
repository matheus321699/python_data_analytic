```python
import pandas as pd
import matplotlib.pyplot as plt

table = pd.read_excel('data_base/gastos_trimestrais_fornecedores.xlsx')
table
```

```python
plt.figure(figsize=(12,6))
table = table.dropna()
grafico_de_barras = plt.bar(table['Descrição'].astype(str), table['Total Anual (R$)'], color='blue')
plt.bar_label(grafico_de_barras, fmt="%.01f", size=10, label_type="edge")
plt.show()
```
